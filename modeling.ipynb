```python
import pandas as pd

df = pd.read_excel('product_sentiment.xlsx')
df.head()
```

```python
X = df['review']
y = df['class']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

### Cross Validation
**KNN**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline

# Define pipeline
pipes = [
    ('vectorizer', TfidfVectorizer()),
    ('model', KNeighborsClassifier())
]
model = Pipeline(pipes)

# Define the parameter grid for grid search
param_grid = {
    'vectorizer__max_features': [1000, 2000, 3000, 5000],  # Experiment with different max_features
    'model__n_neighbors': range(1,10),  # Try different neighbor counts
    'model__metric': ['euclidean', ' minkowski', 'manhattan', 'cosine']  # Consider various distance metrics
}

# Create a GridSearchCV object
grid_search = GridSearchCV(model, param_grid, cv=10)  # Use 10-fold cross-validation

# Fit the grid search to your training data
grid_search.fit(X_train, y_train)
print(grid_search.score(X_test, y_test))

# The best parameter
print(grid_search.best_params_)
```

**SVM**

```python
from sklearn.svm import SVC
from sklearn import svm

# Define pipeline
pipes_svm = [
    ('vectorizer', TfidfVectorizer()),
    ('model_svm', svm.SVC())
]
model_svm = Pipeline(pipes_svm)

# Define the parameter grid for grid search
param_svm = {
    'vectorizer__max_features': [1000, 2000, 3000, 5000],  # Experiment with different max_features
    'model_svm__kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'model_svm__gamma': range(int(0.01 * 100), 10)
}

# Create a GridSearchCV object
grid_svm = GridSearchCV(model_svm, param_svm, cv=10)  # Use 10-fold cross-validation

# Fit the grid search to your training data
grid_svm.fit(X_train, y_train)
print(grid_svm.score(X_test, y_test))

# The best parameter
print(grid_svm.best_params_)
```

**Neural Network**

```python
from sklearn.neural_network import MLPClassifier

# Define pipeline
pipes_mlp = [
    ('vectorizer', TfidfVectorizer()),
    ('model_mlp', MLPClassifier())
]
model_mlp = Pipeline(pipes_mlp )

# Define the parameter grid for grid search
param_mlp = {
    'vectorizer__max_features': [1000, 2000, 3000, 5000],  # Experiment with different max_features
    'model_mlp__activation': ['tanh', 'sigmoid', 'relu'],
}

# Create a GridSearchCV object
grid_mlp = GridSearchCV(model_mlp, param_mlp, cv=10)  # Use 10-fold cross-validation
 
# Fit the grid search to your training data
grid_mlp.fit(X_train, y_train)
print(grid_mlp.score(X_test, y_test))

# The best parameter
print(grid_mlp.best_estimator_)
```

### Modeling
**KNN**

```python
pipes = [
    ('vectorizer', TfidfVectorizer(
        max_features=2000,
    )),
    ('model', KNeighborsClassifier(
        n_neighbors = 8,
        metric = 'cosine',
    )),
]

model = Pipeline(pipes)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

**SVM**

```python
pipes2 = [
    ('vectorizer', TfidfVectorizer(
        max_features=1000,
    )),
    ('model', svm.SVC(
        kernel='linear',
        gamma=1,
    )),
]

model_svm = Pipeline(pipes2)
model_svm.fit(X_train, y_train)
```

```python
model_svm.score(X_test, y_test)
```

**Neural Network**

```python
pipes3 = [
    ('vectorizer', TfidfVectorizer(
        max_features=3000,
    )),
    ('model', MLPClassifier()),
]

model_mlp = Pipeline(pipes3)
model_mlp.fit(X_train, y_train)
```

```python
model_mlp.score(X_test, y_test)
```
